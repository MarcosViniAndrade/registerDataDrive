```python
import pandas as pd
import json
```

```python
file = open('data/pypefy_return_expected.json')
data_json = json.load(file)
file.close()
```

```python
data = []
for node in data_json["data"]["cards"]["edges"]:
    dicts = {}
    for item in node["node"]["fields"]:
        dicts[item["name"]] = item["value"]
    data.append(dicts)

df = pd.DataFrame(data)

df = df.fillna("")

df
```
